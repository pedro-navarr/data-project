```python
%sql
CREATE SCHEMA IF NOT EXISTS Workspace.projeto_analytics_2;
```

```python
df_aeronave = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("sep", ";") \
    .option("encoding", "ISO-8859-1") \
    .load("/Volumes/workspace/projeto_analytics_2/volume_teste/aeronave.csv")

display(df_aeronave.limit(10))
```

```python
df_fator = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("sep", ";") \
    .option("encoding", "ISO-8859-1") \
    .load("/Volumes/workspace/projeto_analytics_2/volume_teste/fator_contribuinte.csv")

display(df_fator.limit(10))
```

```python
df_ocorrencia = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("sep", ";") \
    .option("encoding", "ISO-8859-1") \
    .load("/Volumes/workspace/projeto_analytics_2/volume_teste/ocorrencia.csv")

display(df_ocorrencia.limit(10))
```

```python
df_aeronave.write.format("delta").mode("overwrite").saveAsTable("workspace.projeto_analytics_2.aeronautica_aeronave")

df_fator.write.format("delta").mode("overwrite").saveAsTable("workspace.projeto_analytics_2.aeronautica_fator")

df_ocorrencia.write.format("delta").mode("overwrite").saveAsTable("workspace.projeto_analytics_2.aeronautica_ocorrencia")
```
