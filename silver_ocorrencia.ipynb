```python
meu_schema = "workspace.projeto_analytics_2.aeronautica_ocorrencia"

df_ocorrencia_silver = spark.read.table(meu_schema)

#display(df_ocorrencia_silver.limit(20))
```

```python
df_ocorrencia_silver = df_ocorrencia_silver.drop("codigo_ocorrencia1", "codigo_ocorrencia2", "codigo_ocorrencia3", "codigo_ocorrencia4")

#display(df_ocorrencia_silver)
```

```python
from pyspark.sql import functions as f

colunas = ["ocorrencia_classificacao",
"ocorrencia_latitude",
"ocorrencia_longitude",
"ocorrencia_cidade",
"ocorrencia_uf",
"ocorrencia_pais",
"ocorrencia_aerodromo", "ocorrencia_hora",
"investigacao_aeronave_liberada",
"investigacao_status",
"divulgacao_relatorio_numero",
"divulgacao_relatorio_publicado",
"divulgacao_dia_publicacao"]

dados_sujos = ["NULL", "***", " "]

for c in colunas:
    df_ocorrencia_silver = df_ocorrencia_silver.withColumn(c, f.when(f.col(c).isin(dados_sujos), f.lit(None)).otherwise(f.col(c)))

df_ocorrencia_silver = df_ocorrencia_silver.fillna("NAO INFORMADO")

#display(df_ocorrencia_silver)
```
