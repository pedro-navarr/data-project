```python
meu_schema = "workspace.projeto_analytics_2.aeronautica_fator"
df_fator_silver = spark.read.table(meu_schema)

#display(df_fator_silver.limit(30))
```

```python
from pyspark.sql import functions as f

dados_sujos = ["***", "NULL" " "]
colunas = ["fator_nome", "fator_aspecto", "fator_condicionante", "fator_area"]

for c in colunas:
  df_fator_silver = df_fator_silver.withColumn(c, f.when(f.col(c).isin(dados_sujos), f.lit(None)).otherwise(f.col(c)))

df_fator_silver = df_fator_silver.fillna("NAO INFORMADO")

#display(df_fator_silver.limit(30))
```

```python
df_fator_silver.write.format("delta").mode("overwrite").saveAsTable("workspace.projeto_analytics_2.aeronautica_fator_silver")
```
